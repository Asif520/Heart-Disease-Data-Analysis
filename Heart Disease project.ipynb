## If the cholesterol is high yet oldpeak is low then what are the chances of an individual to have heart attack.


```python
chances_heart = df[(df["chol"] > 230) & (df["oldpeak"] <2.0)]
```

```python
chances_heart
```

```python
chances_heart["target"].value_counts()
```

```python
chances_heart.target[chances_heart.target == 1] = "No"
```

```python
chances_heart.target[chances_heart.target == 0] = "Yes"
```

```python
chances_heart
```

```python
chances_heart["target"].value_counts()
```

```python
chances_heart["target"].value_counts().plot(kind="pie", title="Chances of Heart Attack if Cholesterol High though Old Peak is low")
```
